## Задание 1

```python
# Импорт библиотек
import pandas as pd

# Считывание данных
data = pd.read_csv("laptop_price.csv")
```

```python
# 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU
data['Cpu_Company'] = data['Cpu'].str.split().str[0]
data.head()
```

```python
# 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя
data['Memory_Amount'] = data['Memory'].str.split().str[0]
data.head()
```

```python
# 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)
data['Memory_Type'] = data['Memory'].str.split().str[1:].str.join(' ')
data.head()
```

```python
# 1.4 Удалите признаки Memory и ScreenResolution
data.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
data.head()
```

## Задание 2

```python
# 2.1 Создайте признак SSD, который изначально равен 0
data["SSD"] = 0
```

```python
# 2.2 Поставьте в признаке SSD 1, если ноутбук действительно с типом носителя SSD
data.loc[data["Memory_Type"] == "SSD", "SSD"] = 1
data.head()
```

```python
# 2.3 Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный
data['Weight'] = data['Weight'].str.replace('kg', '').astype(float)
data.head()
```

## Задание 3
Создайте датафрейм с клиентами.
<br>laptop_id - это индексы датафрейма с ноутбуками

```python
clients = pd.DataFrame({
 'client_id': [45, 32, 67, 33, 43],
 'laptop_id': [506, 398, 710, 120, 1999]
})
```

```python
# 3.1 Присоедините к таблице clients данные по ноутбукам через метод join
clients_join = clients.join(data.set_index('laptop_ID'), on='laptop_id')
clients_join.head()
```

```python
#3.2 Присоедините к таблице clients данные по ноутбукам через метод merge
clients_merge = clients.merge(data, left_on='laptop_id', right_index=True)
clients_merge.head()
```

## Задание 4

```python
# 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя. Отсортируйте от меньшей стоимости к большей
data.groupby("Company")["Price_euros"].mean().sort_values()
```

```python
data.pivot_table(index="Company", values="Price_euros", aggfunc="mean").sort_values(by="Price_euros")
```

```python
# 4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора (Cpu_Company)
data.groupby("Cpu_Company").agg({"Price_euros":["min", "max", "mean"]})
```

```python
data.pivot_table(index="Cpu_Company", values="Price_euros", aggfunc=["min", "max", "mean"])
```

```python
# 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ
data.pivot_table(index='Cpu_Company', columns='Ram', values='Price_euros', aggfunc='count', fill_value=0)
```

```python
data.groupby(["Cpu_Company", "Ram"])["laptop_ID"].count()
```

```python
# 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы (OpSys) и GB памяти (Memory_Amount)
data.pivot_table(index='OpSys', columns='Memory_Amount', values='Price_euros', aggfunc='mean')
```

```python
data.groupby(['OpSys','Memory_Amount' ])['Price_euros'].mean()
```
